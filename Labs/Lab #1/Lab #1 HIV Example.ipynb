```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create array of 101 numbers between 0 and 1

time = np.linspace(0,1,101)
```

```python
# Use equation 5.1
# B = 0

B = 0
A = 100
alpha = 25
beta = 35

viral_load = A*np.exp(-alpha*time) + B*np.exp(-beta*time)
```

```python
plt.plot(time, viral_load,'r')

plt.xlabel('Time')
plt.ylabel('Viral Load')
plt.title('Viral Load vs Time')

plt.show()
```

```python
# Load data
data = np.loadtxt('HIVseries.csv', delimiter=',')

# Put each column into a separate array.
N = np.size(data,0)

time_data = data[0:N:1,0]
virus_concentration = data[0:N:1,1]
```

```python
plt.plot(time_data, virus_concentration,'go',label="Data Points")

plt.legend()

plt.xlabel('Time in days')
plt.ylabel('Viral Load')
plt.title('Concentration of HIV over a week')

plt.show()
```
