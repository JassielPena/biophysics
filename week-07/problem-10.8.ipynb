```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rand = default_rng().random
```

```python
Simulations = 150

lini = 0
beta_birth = 1.5/60
k_death = 0.014/60

N =1000

Time = np.zeros(Simulations)    
TrialsZero = np.zeros(Simulations)

for i in range(Simulations):
    l = lini
    T = 0
    Count = 0

    TS = np.zeros(N)
    LS = np.zeros(N)

    for n in range(N):
        rate = beta_birth + k_death * l
        t = - 1/rate * np.log(rand())
        xi = beta_birth / rate
        event = rand() < xi
        
        if event:
            l = l+1
        else:
            l = l-1
        
        T = T+t

        TS[n] = T
        LS[n] = l
    
    #Count number of trials where l = 0
    Data = LS[LS <= 0]
    TimeData = TS[LS <= 0]
    CountZero = Data.size
    
    if Data.size == 0:
        Time[Count] = 0
        TrialsZero[Count] = CountZero
        Count += 1
            
    else:
        Time[Count] = Data[-1]
        TrialsZero[Count] = CountZero
        Count += 1

plt.figure(dpi=300)

plt.plot(Time, TrialsZero, 'ro');
```
