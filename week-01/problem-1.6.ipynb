```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1.6 (a)

```python
#Cat, gopher, human, elephant, horse

# Body Mass (Kg)
BM = [4.5, 0.23, 73, 5500, 465]

# Approx. genome size (million basepairs)
Size = [3070, 3270, 3455, 4220, 3178]
```

```python
plt.plot(BM, Size,'ro')

plt.xlabel('Body Mass (Kg)')
plt.ylabel('Approx. genome size (million basepairs)')
plt.title('Genome size vs Body Mass')

plt.show()
```

Is there a relation between these two variables?

It looks like the size of the Genome is proportional to the body mass of the animal. It looks like a linear relationship. The larger the Body Mass, the bigger the genome size.

## 1.6 (b)


```python
# poison dart frog
BM_frog = 0.003
Size_frog = 8830
```

```python
plt.plot(BM, Size,'ro', label="Original Data Set")
plt.plot(BM_frog, Size_frog,'bo', label="Poison Dart Frog")

plt.legend()

plt.xlabel('Body Mass (Kg)')
plt.ylabel('Approx. genome size (million basepairs)')
plt.title('Genome size vs Body Mass')

plt.show()
```

What happened to the relation you previously found? 
Did the first five points have something in common that made them not representative of all animals?

The frog does not follow that same relationship. It has a small body mass but a much larger genome size.

The first five points were all mammals. Not all animals are mammals so that means the first five points were not a good representation of all animals.
