## 5.1 HIV Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create array of 101 numbers between 0 and 1

time = np.linspace(0,1,101)
```

```python
# Use equation 5.1
# B = 0

B = 0
A = 100
alpha = 25
beta = 35

viral_load = A*np.exp(-alpha*time) + B*np.exp(-beta*time)
```

```python
plt.plot(time, viral_load,'r')

plt.xlabel('Time')
plt.ylabel('Viral Load')
plt.title('Viral Load vs Time')

plt.show()
```

```python
# Additional Plots

B1 = 20
A1 = 0
alpha1 = 15
beta1 = 225

viral_load1 = A1*np.exp(-alpha1*time) + B1*np.exp(-beta1*time)


B2 = 75
A2 = 50
alpha2 = 0
beta2 = 25

viral_load2 = A2*np.exp(-alpha2*time) + B2*np.exp(-beta2*time)


B3 = 77
A3 = 90
alpha3 = 13
beta3 = 0

viral_load3 = A3*np.exp(-alpha3*time) + B3*np.exp(-beta3*time)

B4 = 1690
A4 = 1400
alpha4 = 34
beta4 = 77

viral_load4 = A4*np.exp(-alpha4*time) + B4*np.exp(-beta4*time)


plt.plot(time, viral_load,'r',label="Run 1")
plt.plot(time, viral_load1,'b',label="Run 2")
plt.plot(time, viral_load2,'c',label="Run 3")
plt.plot(time, viral_load3,'y',label="Run 4")
plt.plot(time, viral_load1,'k',label="Run 5")

plt.legend()

plt.xlabel('Time')
plt.ylabel('Viral Load')
plt.title('Viral Load vs Time')

plt.show()
```

```python
# Load data
data = np.loadtxt('HIVseries.csv', delimiter=',')

# Put each column into a separate array.
N = np.size(data,0)

time_data = data[0:N:1,0]
virus_concentration = data[0:N:1,1]
```

```python
plt.plot(time_data, virus_concentration,'go',label="Data Points")

plt.legend()

plt.xlabel('Time in days')
plt.ylabel('Viral Load')
plt.title('Concentration of HIV over a week')

plt.show()
```

(a) 
Adjust four parameters until model and data appear in the plot

```python
Time = np.linspace(time_data.min(),time_data.max(),101)
B = 85000
A = 85000
alpha = .4
beta = 2

viral_load = A*np.exp(-alpha*Time) + B*np.exp(-beta*Time)

plt.plot(Time, viral_load,'r', label="Model")
plt.plot(time_data, virus_concentration,'go',label="Data Points")

plt.legend()

plt.xlabel('Time in days')
plt.ylabel('Viral Load')
plt.title('Concentration of HIV over a week')

plt.show()
```

(b)

I started by adjusting beta. I observed that this mainly changed the way the first half of the model. I used to fit the first part of the data (0 - 3 days). Once I felt it was a good fit then I began to change alpha. This affected the second half of the model. I kept changing it until I felt I had a good fit for the entire data set.

(c)
How does the inverse of the T-cell infection rate compare to the latency period

```python
# alpha = rate at which T-cells are infected

Time_Latency = np.linspace(0.0001,3600,10000)


inverse_alpha = 1/Time_Latency

plt.plot(Time_Latency,inverse_alpha ,'b', label="Inverse alpha")

plt.legend()

plt.xlabel('Time in days')
plt.ylabel('1/Viral Load')
plt.title('Inverse alpha')

plt.show()
```

The model suggests that suggests that HIV has a latency period because initially beta is larger than alpha. The virions are being removed at a faster rate than new T-cells are being infected.
