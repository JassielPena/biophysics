```python
import scipy
import numpy as np
import matplotlib.pyplot as plt

# Create random number generator
from numpy.random import default_rng
rand = default_rng().random

from scipy.special import factorial
```

```python
l_values = np.linspace(0,100,100)

P_l = (np.exp(-8) * 8**l_values)/factorial(l_values)

plt.figure(dpi=300)

plt.plot(l_values, P_l);

plt.title("Poison Distribution",fontsize=12);
plt.ylabel("Probability", fontsize=10);
plt.xlabel("Input Value", fontsize=10);
```

```python
# heads = 1 and tails = -1

Ntrials = 1000
Nflips = 100

data = np.zeros(Ntrials)
count = 0

# Generate Ntrials sequences of Nflips coin flips, where heads comes up 8% of the time
for n in range(Ntrials):
    flips = 2*(rand(Nflips) <= 0.08) - 1
    
    Heads = 100 + flips.sum()
    
    data[count] = Heads
    count += 1

if count == 0:
    print("No trials started had at least 1 head.")
    
plt.figure(dpi=300)

plt.hist(data,bins='auto');

plt.title("Fruquency of Heads",fontsize=12)
plt.ylabel("Frequency", fontsize=10)
plt.xlabel("Value", fontsize=10)
```

```python
# heads = 1 and tails = -1

Ntrials = 1000000
Nflips = 100

data = np.zeros(Ntrials)
count = 0

# Generate Ntrials sequences of Nflips coin flips, where heads comes up 8% of the time
for n in range(Ntrials):
    flips = 2*(rand(Nflips) <= 0.08) - 1
    
    Heads = 100 + flips.sum()
    
    data[count] = Heads
    count += 1

if count == 0:
    print("No trials started had at least 1 head.")

# Poisson Distribution * N

l_values = np.linspace(0,200,100)
P_l = (np.exp(-8) * 8**l_values)/factorial(l_values)

PlxN = P_l * Ntrials

plt.figure(dpi=300)

plt.hist(data,bins='auto');
plt.plot(l_values, PlxN);

plt.title("Fruquency of Heads",fontsize=12)
plt.ylabel("Frequency", fontsize=10)
plt.xlabel("Value", fontsize=10)

plt.show()
```
