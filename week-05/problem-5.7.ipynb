```python
import numpy as np
import matplotlib.pyplot as plt

# Create random number generator
from numpy.random import default_rng
rand = default_rng().random
```

## (a) __Histogram of $1/x$__

```python
#10,000 random numbers between 0 and 1

Trials = 10000

X = rand(Trials)

X_Inverse = 1/X

XI_max = X_Inverse.max()
XI_min = X_Inverse.min()

plt.figure(dpi=300)

counts, bin_edges, _ = plt.hist(X_Inverse, bins=100, range=(XI_min,XI_max));

plt.title("Inverse of 10,000 random numbers between 0 and 1", fontsize=12)
plt.ylabel("Frequency", fontsize=10)
plt.xlabel("Numbers", fontsize=10)

print(XI_min,XI_max)
```

## (b) Log-log plot of frequencies

```python
offset = 0.5*(bin_edges[1] - bin_edges[0])
plt.figure(dpi=300)
plt.loglog(bin_edges[:-1] + offset, counts)

plt.title("Log-log of Frequencies", fontsize=12)
plt.ylabel("Frequency", fontsize=10)
plt.xlabel("Numbers", fontsize=10)
```

## (c) Analysis and Conclusions

For the most part it does appear to look like a power-law distribution. The number of bins used does affect the appearance of the figures. The more bins you use then the more random the tail end of the figues looks. As x gets closer to 0 then the inverse gets closer to infinity. The range of the figure goes from 1 to infinity. There are inifinite possible numbers between 0 and 1 therefore you are likely to get a lot of variance between each number. 1/x should produce this type of graph. As x gets smaller it approaches inifinty, and as x gets larger it should approach 0.
