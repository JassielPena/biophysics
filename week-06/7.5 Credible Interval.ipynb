```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import beta
```

## a.

```python
M = 600

l = 301

B = 1/beta(l + 1, M-l + 1)

xi = np.linspace(0,1,1000)

P = B * xi**l * (1-xi)**(M-l)

plt.plot(xi,P)
```

```python
dxi = xi[1] - xi[0]

Mxi = l/M
Y = .0335
P[(xi > (Mxi-Y)) * (xi < (Mxi+Y))].sum() * dxi
```
