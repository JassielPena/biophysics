```python
# Load hemoglobin data
data = np.loadtxt('hemoglobin.csv', delimiter=',')

Concentration = data[:,0]
ProbUnbound = data[:,1]

C = np.linspace(0 , .00003, 1001)
Kd = .001

C_1 = C/Kd
n = .45

ProbUnbound_Model = 1 / (1 + C_1**n)

plt.plot(Concentration, ProbUnbound, '.', label='Hemoglobin Data');
plt.plot(C, ProbUnbound_Model, 'g', label='Hemoglobin Model');
plt.xlabel('Concetration [M]');
plt.ylabel('Probability to be unbound');
plt.legend();
```

## (b) Least-squares fit

```python
X = Concentration
Y = ProbUnbound

Mean_X = np.mean(X)
Mean_Y = np.mean(Y)

Number = len(X)

Numerator = 0
Denominator = 0

for i in range(Number):
    Numerator += (X[i] - Mean_X) * (Y[i] - Mean_Y)
    Denominator += (X[i] - Mean_X)**2
    
M = Numerator / Denominator

Intercept = Mean_Y - (M * Mean_X)

LeastSquare = M*C + Intercept

plt.plot(Concentration, ProbUnbound, '.', label='Hemoglobin Data');
plt.plot(C, ProbUnbound_Model, 'g', label='Hemoglobin Model - Guess');
plt.plot(C, LeastSquare, 'r', label='Hemoglobin Model- Least Squares');
plt.xlabel('Concetration [M]');
plt.ylabel('Probability to be unbound');
plt.legend();
```
