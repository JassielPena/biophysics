# 11.5 PENDULUM PHASE PORTRAIT

```python
import numpy as np
import matplotlib.pyplot as plt
```

## (a) Plot the phase portrait.

```python
x = np.linspace(-1,1, 21)
y = np.linspace(-1,1, 21)

X,Y = np.meshgrid(x,y)

vX = Y
vY = -X
plt.quiver(X,Y,vX,vY, pivot='mid');
plt.plot(x,x**3);
```

## (b) Describe the motion.

The lines circle around the center, the radius decreases as you move forward.
